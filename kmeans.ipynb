```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
points=np.array([
    [1,2],
    [1,4],
    [1,0],
    [5,2],
    [5,4],
    [6,2]
])

labels=["A","B","C","D","E","F"]
```

```python
#kmeans algo
kmeans=KMeans(n_clusters=2,random_state=42,n_init=10)
kmeans.fit(points)
clusters=kmeans.labels_
centroids=kmeans.cluster_centers_
```

```python
#results
print("cluster assistant")
for i,label in enumerate(labels):
    print(f"point{label} {points[i]} -> cluster {clusters[i]}")

print("\n final centriods")
print(centroids)
```

```python
#visualization
plt.figure(figsize=(6,6))
plt.scatter(points[:,0],points[:,1],c=clusters, s=100,cmap="viridis",label="points")


plt.scatter(centroids[:,0],centroids[:,1],c='red',marker='X',s=200,label="centroids")
for i,txt in enumerate(labels):
    plt.annotate(txt,(points[i,0]+0.1,points [i,1]+0.1))
    

plt.title("kmeans clustering")
plt.ylabel("x2")
plt.xlabel("x1")
plt.legend()
plt.grid(True)
plt.show()
```
